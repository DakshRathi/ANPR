```python
from ultralytics import YOLO
```

```python
# Load a model
model = YOLO("yolo11s.pt")

# Train the model
train_results = model.train(
    data="/Users/daksh/Desktop/ANPR/datasets/dataset.yaml",  # path to dataset YAML
    freeze=[0, 1],
    epochs=40,  # number of training epochs
    imgsz=480,  # Reduced training image size
    device='mps', 
    batch=16,  # Reduced batch size
    amp=True,  # Enable mixed precision
)
```

```python
# Evaluate model performance on the validation set
metrics = model.val()

# Perform object detection on an image
results = model("/Users/daksh/Desktop/ANPR/datasets/yolo/data/images/test/video11_4950.jpg")
results[0].show()
```
